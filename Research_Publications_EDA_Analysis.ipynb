## 3. Data Preprocessing & Cleaning <a id="cleaning"></a>

Before proceeding to analysis, we conducted a rigorous data quality check.
The dataset was found to be highly structured with no missing values or duplicates.

**Actions Taken:**
1.  **Column Renaming:** Renamed columns to follow standard naming conventions (snake_case) for better readability.
2.  **Precision Formatting:** Rounded off numerical columns to 2 decimal places to maintain consistency.
3.  **Outlier Finding** Detected Outliers from each columns

```python
# 1. Check for missing values
missing_count = df.isnull().sum().sum()
print(f"Total Missing Values: {missing_count}")

# 2. Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Total Duplicates: {duplicate_count}")
```

```python
# Renamed columns for better readability
df.rename(columns={
    'Category Normalized Citation Impact': 'CNCI',
    'year': 'Year'
}, inplace=True)

# Rounding off decimal columns to 2 decimal places 
df = df.round({
    'Collab-CNCI' : 2,
    'CNCI' : 2
})

df.head(2)
```

```python
#Checking Outliers using Box Plot
numerical_cols = df.select_dtypes(include=['number']).columns #select numberical columns

# Loop on numberical columns to plot boxplot
for col in numerical_cols:
    fig = px.box(df, y=col, title=f"Outlier Check for {col}")
    fig.show()
```

```python
#Checking Outliers using IQR
col_name = 'Times Cited'

Q1 = df[col_name].quantile(0.25)
Q3 = df[col_name].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Wo rows dhoondhein jo outliers hain
outliers = df[(df[col_name] < lower_bound) | (df[col_name] > upper_bound)]

print(f"No. of Outliers in {col_name} is {len(outliers)}")
outliers
```
