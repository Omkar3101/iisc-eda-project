## 3. Data Preprocessing & Cleaning <a id="cleaning"></a>

Before proceeding to analysis, we conducted a rigorous data quality check.
The dataset was found to be highly structured with no missing values or duplicates.

**Actions Taken:**
1.  **Column Renaming:** Renamed columns to follow standard naming conventions (snake_case) for better readability.
2.  **Precision Formatting:** Rounded off numerical columns to 2 decimal places to maintain consistency.
3.  **Outlier Finding** Detected Outliers from each columns

```python
# 1. Check for missing values
missing_count = df.isnull().sum().sum()
print(f"Total Missing Values: {missing_count}")

# 2. Check for duplicates
duplicate_count = df.duplicated().sum()
print(f"Total Duplicates: {duplicate_count}")
```

```python
# Renamed columns for better readability
df.rename(columns={
    'Category Normalized Citation Impact': 'CNCI',
    'year': 'Year'
}, inplace=True)

# Rounding off decimal columns to 2 decimal places 
df = df.round({
    'Collab-CNCI' : 2,
    'CNCI' : 2
})

df.head(2)
```

```python
#Checking Outliers using Box Plot
numerical_cols = df.select_dtypes(include=['number']).columns #select numberical columns

# Loop on numberical columns to plot boxplot
for col in numerical_cols:
    fig = px.box(df, y=col, title=f"Outlier Check for {col}")
    fig.show()
```

```python
#Checking Outliers using IQR
col_name = 'Times Cited'

Q1 = df[col_name].quantile(0.25)
Q3 = df[col_name].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Wo rows dhoondhein jo outliers hain
outliers = df[(df[col_name] < lower_bound) | (df[col_name] > upper_bound)]

print(f"No. of Outliers in {col_name} is {len(outliers)}")
outliers
```

### 4.1 Univariate Analysis
*Understanding the distribution of individual variables.*

```python
# Numerical columns (Numbers: int, float)
num_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Categorical columns (Text, Categories)
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

print("Numerical Columns:", num_cols)
print("Categorical Columns:", cat_cols)
```

### Numerical Columns

```python
#Statistical Summary of Numerical Columns
df.describe()
```

```python
#Histogram and KDE 

# Loop through columns
for col in num_cols:
    
    if col == 'Year': 
        continue

    # 1. Clean Data
    data = df[col].dropna()
    
    if len(data) == 0:
        continue
        
    # 2. Statistics Calculation
    mean_val = data.mean()
    median_val = data.median()
    std_val = data.std()
    
    # 3. KDE Curve Calculation
    x_range = np.linspace(data.min(), data.max(), 500)
    kde = stats.gaussian_kde(data)
    y_kde = kde(x_range)
    
    # 4. Figure Setup
    fig = make_subplots(specs=[[{"secondary_y": True}]])
    
    # Histogram
    fig.add_trace(go.Histogram(
        x=data,
        name='Count',
        marker_color='#a3c2fa', 
        opacity=0.5,
        nbinsx=100
    ), secondary_y=False)
    
    # KDE Line
    fig.add_trace(go.Scatter(
        x=x_range,
        y=y_kde,
        mode='lines',
        name='KDE Density',
        line=dict(color='#00008b', width=3)
    ), secondary_y=True)
    
    #Zone Creation    
    # Zone 3: 99.7% (Mean ¬± 3 SD)
    fig.add_vrect(
        x0=max(0, mean_val - 3*std_val), x1=mean_val + 3*std_val,  # use max(0, calculation) to not show negative x-axis
        fillcolor="orange", opacity=0.05, layer="below", line_width=0,
        annotation_text="99.7%", annotation_position="top right"
    )
    # Zone 2: 95% (Mean ¬± 2 SD)
    fig.add_vrect(
        x0=max(0, mean_val - 2*std_val), x1=mean_val + 2*std_val,
        fillcolor="orange", opacity=0.1, layer="below", line_width=0,
        annotation_text="95%", annotation_position="top right"
    )
    # Zone 1: 68% (Mean ¬± 1 SD)
    fig.add_vrect(
        x0=max(0, mean_val - std_val), x1=mean_val + std_val,
        fillcolor="orange", opacity=0.2, layer="below", line_width=0,
        annotation_text="68%", annotation_position="top right"
    )

    # Standard Deviation Labels
    sd_labels = [-3, -2, -1, 1, 2, 3] 
    
    for sigma in sd_labels:
        sd_point = mean_val + (sigma * std_val)
        
        if sd_point > 0:
            #Draw a thin dotted line
            fig.add_vline(x=sd_point, line_width=1, line_dash="dot", line_color="gray", opacity=0.5)
            
            #Add text label 
            fig.add_annotation(
                x=sd_point,
                y=1,
                yref="paper",
                text=f"{sigma:+}œÉ", 
                showarrow=False,
                font=dict(size=12, color="#555"),
                yshift=15 
            )

    # Mean & Median Lines
    fig.add_vline(x=mean_val, line_width=3, line_dash="dash", line_color="red", 
                  annotation_text=f"Mean: {mean_val:.2f}", annotation_position="top right")
    
    fig.add_vline(x=median_val, line_width=3, line_dash="dot", line_color="green", 
                  annotation_text=f"Median: {median_val:.2f}", annotation_position="bottom left")

    # Layout
    fig.update_layout(
        title=f'Distribution of {col} (SD={std_val:.2f})',
        bargap=0.2, 
        height=600,
        showlegend=True,
        template='plotly_white'
    )
    
    fig.update_yaxes(title_text="<b>Count</b>", secondary_y=False)
    fig.update_yaxes(title_text="<b>KDE Density</b>", secondary_y=True, showgrid=False)
    fig.update_xaxes(title_text=col)

    fig.show()
```
